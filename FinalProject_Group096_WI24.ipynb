## Background and Prior Work

Introduction:
	Airbnb is an online marketplace which manages short-term homestays through homeowners. Users can scroll through listings and rent out a home to stay at for a certain period of time. A cut of the money goes to the company as commission. As traveling becomes more popular through social media and remote jobs, Airbnb also becomes more prevalent and widespread among the general public. It is estimated that every second, six guests check into a listing (Woodward, AIRBNB STATISTICS [2024]: USER & MARKET GROWTH DATA)<a name="cite_ref-3"></a>[<sup>3</sup>](#cite_note-3). By observing datasets about Airbnb listing prices, we can make inferences about the future prices of current Airbnb listings and any potential changes in them. This knowledge can provide hosts and users a more transparent and efficient experience, as well as ensure hosts are not price gouging their listings.

We will also only be looking at 7 variables: Property Type, Room Type, Number of bedrooms, Number of bathrooms, number of accommodates, Neighborhood Cleansed and Neighborhood Group Cleansed.

## Why these variables? We feel like these variables would contribute the most because:
1. `'Property Type'`: This refers to the kind of property being listed, such as an apartment, house, condominium, loft, etc. The property type is important because different types of properties usually have different market values and appeal to different kinds of guests. For instance, a house might be more appealing to large groups or families which can lead to a higher price as there are more people.

2. `'Room Type'`: This variable includes 'Entire home/apt', 'Private room', 'Shared room. Room type directly affects the price because entire homes or apartments typically cost more than private or shared rooms, reflecting the level of privacy and space offered to guests.

3. `'Bedrooms'`: The number of bedrooms is also a useful variable in predicting a listing price because it often correlates with the accommodation's capacity to host guests. In addition, more bedrooms can accommodate more guests, which increases the price due to the fact that there is more people.

4. `'Bathrooms'`: The number of bathrooms also influences the price of the listing because bathrooms also contribute the amount of space you rent. So the more bathrooms a listing has, the more space there is in the listing, which increases the price.

5. `'Accommodates'`: This refers to the maximum number of guests a listing can accommodate. Listings that can accommodate more guests are typically priced higher due to the flexibility the listing offers. This variable is related to the number of bedrooms but provides a more exact measurement.

6. `'Neighborhood Group Cleansed'`: The neighborhood group where the listing is located can impact the price due to factors like accessibility, safety, proximity to transit, etc. For instance, listings in Manhattan might be more expensive due to the popularity among tourists.

7. `'Neighborhood Cleansed'`: Within each neighborhood group, specific neighborhoods can have vastly different characteristics and appeal, further refining the price prediction. For example, within Brooklyn, a listing in Williamsburg might be priced differently from one in Bay Ridge due to differences in popularity, accessibility, and activites.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import OneHotEncoder
```

## New York City Airbnb Listings

### Importing the dataset

```python
nyc_airbnb= pd.read_csv('airbnb-listings.csv', delimiter=";", low_memory=False)
nyc_airbnb.head()
```

## Getting Necessary Columns

```python
nyc_airbnb = nyc_airbnb[['Property Type', 
                         'Room Type', 
                         'Bedrooms',
                         'Bathrooms', 
                         'Accommodates',
                         'Neighbourhood Cleansed',
                         'Neighbourhood Group Cleansed',
                        'Price']]
nyc_airbnb.dropna(subset=['Price'], inplace=True) # Drops rows where price is missing
nyc_airbnb.head()
```

### One Hot Encode for `'Property Type'`, `'Room Type'`,  `'Neighbourhood Cleansed'`, and `'Neighbourhood Group Cleansed'`

```python
categorical_cols = ['Property Type', 'Room Type', 'Neighbourhood Group Cleansed', 'Neighbourhood Cleansed']

encoder = OneHotEncoder(sparse=False)

encoded_columns = encoder.fit_transform(nyc_airbnb[categorical_cols])

encoded_columns_df = pd.DataFrame(encoded_columns, columns=encoder.get_feature_names_out(input_features=categorical_cols))

nyc_airbnb_dropped = nyc_airbnb.drop(categorical_cols, axis=1)

nyc_airbnb_encoded= pd.concat([nyc_airbnb_dropped, encoded_columns_df], axis=1)

nyc_airbnb_encoded.dropna(subset=['Price'], inplace=True) #Drops rows where price is N/A
nyc_airbnb_encoded.head()
```

# Results

## Exploratory Data Analysis on New York City Airbnb Listings

### Relationship between Room Type and Prices in Airbnb Listings

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=nyc_airbnb, y='Room Type', x='Price', palette='coolwarm')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.title('Box Plot of NYC Airbnb Prices by Room Type', fontsize=18)
plt.xlabel('Price ($)', fontsize=14)
plt.ylabel('Room Type', fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Based off of this graph, we can easily see that listings that offer the entire house and apartments have a higher variability than both private rooms and shared rooms largely due to the fact that there are more variables to consider when renting an entire home or apartment like the amount of bedrooms, bathrooms, etc. in comparision to a private/shared room. As we predicted, renting an entire home and apartment has a higher median price point, followed by private rooms, and then shared rooms. This makes sense intuitively because the more privacy and space given will generally increase the price of the listing. Entire homes/apartments and private rooms have the most amoiunt of outliers which is likely due to the fact that there are high-demand areas and more desirable places, which leads to a higher price listing in their respective room types.

### Relationship between Airbnb Listing Prices and Neighbourhood Groups in New York City


```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=nyc_airbnb, y='Neighbourhood Group Cleansed', x='Price', palette='coolwarm')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.title('Box Plot of NYC Airbnb Prices by Neighbourhood Group', fontsize=18)
plt.xlabel('Price ($)', fontsize=14)
plt.ylabel('Neighbourhood Group', fontsize=14)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

As we can see through the different neighbourhood groups, Manhattan has the highest price range and median price in comparison to the other groups. This is likely due to the fact that Manhattan is one of the highest destination spots in New York City amongst tourists. However, as we move away from Manhattan and towards the outer neighbourhoods like the Bronx and Queens, the median price and overall price range of renting rooms there are cheaper than other neighbourhoods. In addition, the spread of data points for Manhattan and Brooklyn suggests a greater diversity in the types of listings available, from budget to premium, while the other neighbourhoods seem to cater more to the lower to mid-range market.

### Distribution of Bedrooms in NYC Airbnb Listings

```python
plt.figure(figsize=(12, 8))
sns.countplot(x='Bedrooms', data=nyc_airbnb, palette='viridis')
sns.set_style("whitegrid")
plt.title('Distribution of NYC Airbnb Listings by Number of Bedrooms', fontsize=18)
plt.xlabel('Number of Bedrooms', fontsize=14)
plt.ylabel('Count of Listings', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Generally, we can expect most of the bedrooms in our dataset to have only 1 as New York City is very small and is hard to find apartments/houses with more than 1 bedroom. As to why there are 0 bedrooms in the listings, it is likely due to the fact that many of the "apartments" are studios. These are typically one large room that serves as the living, dining, and bedroom combined. Since there isn't a separate bedroom, these listings are often classified as having "0 bedrooms". It could also be the fact that the listing is for a shared space instead of a single private room/apartment.

### Distribution of Bathrooms in NYC Airbnb Listings

```python
plt.figure(figsize=(14, 8))
sns.countplot(x='Bathrooms', data=nyc_airbnb, palette='rocket')
sns.set_style("whitegrid")
plt.title('Distribution of Airbnb Listings by Number of Bathrooms', fontsize=20)
plt.xlabel('Number of Bathrooms', fontsize=16)
plt.ylabel('Count of Listings', fontsize=16);
```

Similarly to the distribution of bedrooms, since a majority of the listings for rent are about one bedroom, we can expect the same for the bathroom to be about one as well. As to why there are few zeroes, it is likely due to the fact that shared bathrooms are not tied to a room and can generally be shared with everyone in the apartment, hence no one "owns" the restroom. The 0.5s indicate that while there is a toilet, there is no shower or bathtub in the bathroom. 

### Distribution of Airbnb Room Types in New York City

```python
room_type_counts = nyc_airbnb['Room Type'].value_counts()
plt.figure(figsize=(9, 7))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette("pastel"))
plt.title('Distribution of NYC Airbnb Room Types')
plt.show()
```

Since Airbnbs are meant for home owners to rent properties, it makes sense that a good majority for the listings in New York City being entire homes/apartments. But it was actually very surprising to see a good amount being private rooms as well. This is likely to promote a more budget friendly approach for those who want to go to New York City but do not have enough to rent an entire space. Lastly, there are is a very small percentage of shared rooms throughout the whole dataset likely due to the low demand for these types of listings as many probably do not enjoy sharing the room with others and not willing to sacrifice budget for privacy.
